```python
import numpy as np
import matplotlib.pyplot as plt
from nnfs.datasets import sine_data
# X=np.array([
#     [230.1,37.8,69.2],
#     [44.5,39.3,45.1],
#     [17.2,45.9,69.3],
#     [151.5,41.3,58.5],
#     [180.8,10.8,58.4],
#     [8.7,48.9,75],
#     [57.5,32.8,23.5],
#     [120.2,19.6,11.6],
#     [8.6,2.1,1],
#     [199.8,2.6,21.2]
# ])
# y=np.array([22.1,10.4,9.3,18.5,12.9,7.2,11.8,13.2,4.8,10.6])
# y = y.reshape(-1, 1)
X, y = sine_data()

```

```python
from Dense import Model, Layer_Dense, Activation_Linear, Activation_ReLU, OptimizerAdam, AccuracyRegression, Loss_MeanSquaredError
```

```python
model = Model()
model.add(Layer_Dense(X.shape[1], 128))
model.add(Activation_ReLU())
model.add(Layer_Dense(128, 128))
model.add(Activation_ReLU())
model.add(Layer_Dense(128, 1))
model.add(Activation_Linear())

model.set(
  loss=Loss_MeanSquaredError(),
  optimizer=OptimizerAdam(learning_rate=0.0005, decay=1e-5),
  accuracy=AccuracyRegression()
)


model.finalize()

model.train(X, y, epochs=10000, printEvery=100)
```

```python
X_test, y_test = sine_data()
model.evaluate(X_test, y_test)
```

```python
plt.plot(X_test, y_test)
plt.plot(X_test, model.outputLayerActivation.output)
plt.show()
# print(model.outputLayerActivation.output)
```
